### `Import Libraries`

```python
import numpy as np
import sympy as sym
from numpy import diff
from scipy.misc import derivative
from matplotlib import pyplot as plt
```

### `Sample Data`

```python
x_list = [1,4,5,7,8,10] # x
y_list = [x**2 for x in x_list] # y = f(x) 
print("x_list --> ", x_list)
print("y_list --> ", y_list)
```

- **`Plot`**

```python
plt.plot(x_list,y_list, label='x vs y = f(x**2)')
plt.xlabel("x")
plt.ylabel("y = f(x**2)")
plt.legend()
plt.show()
```

### `1st Differentiation of [y] w.r.t [x]`

```python
dydx = diff(y_list)/diff(x_list)
print("dydx --> ", dydx)
```

### `1st Differentiation of [y] without affect of [x]`

```python
dydx = diff(y_list)
print("dydx --> ", dydx)
```

### `Find Rate of Change[Slope/Derivative] in function x, y = f(x**2)`

```python
x = sym.Symbol('x')
function = x ** 2
derivitive = sym.diff(function)
print(derivitive)
```

- **Observation**
- This means whenever there is a change in x(or at each x value) and which tends to y = f(x**2), will have rate of change in values i.e., 2x.
- In other words it can be known as speed / slope of a functions

```python
def f(x):
    return x ** 2

def derive(function, value):
    h = 0.00000000001
    top = function(value + h) - function(value)
    bottom = h
    slope = top / bottom
    # Returns the slope with 3 decimals
    return float("%.3f" % slope)

print("If x = 2, rate of change[derivative/slope] in y :",derive(f, 2))
print("If x = 4, rate of change[derivative/slope] in y :",derive(f, 4))
print("If x = 10, rate of change[derivative/slope] in y :",derive(f, 10))    
```

- **Observation**
- This values indicates, in case of value increase in x tends to have high more increase(rate of change) in y

### `More Examples of derivatives on different functions `

- `Find Rate of Change[Slope/Derivative] in function x, (y/f(x)) | f(x) = x`
- Common Functions : Line
- Function : f(x) = x
- Derivative : 1

```python
x = sym.Symbol('x')
function = x
derivitive = sym.diff(function)
print(derivitive)
```

- **Observation**
- When both x & y[f(x)] are same, then slope/rate of change will same always be constant. i.e., 1 

```python
def f(v):
    return v
def get_derive(function, value):
    h = 0.00000000001
    top = function(value + h) - function(value)
    bottom = h
    slope = top / bottom
    # Returns the slope with 3 decimals
    return float("%.3f" % slope)

y = f(x_list)
print("x --> ", x_list)
print("y --> ", y)
derivatives = [get_derive(f, value) for value in x_list]
print("derivatives --> ", derivatives)

plt.plot(x_list,y, label='x vs y = f(x)')
plt.plot(x_list,derivatives, label='derivatives')
plt.xlabel("x")
plt.ylabel("y = f(x)")
plt.legend()
plt.show()
```

- **Observation**
- When both x & y[f(x)] are same, then slope/rate of change will same always be constant. i.e., 1 

- `Find Rate of Change[Slope/Derivative] in function x, (y/f(2x)) | f(2x) = x`
- Common Functions : Line
- Function : f(2x) = x
- Derivative : 2

```python
x = sym.Symbol('x')
function = 2 * x
derivitive = sym.diff(function)
print(derivitive)
```

- **Observation**
- When both x & y[f(x)] are same with multiplcation of 2 , then slope/rate of change will same always be constant. i.e., 2 

```python
def f(v):
    return 2 * v
def get_derive(function, value):
    h = 0.00000000001
    top = function(value + h) - function(value)
    bottom = h
    slope = top / bottom
    # Returns the slope with 3 decimals
    return float("%.3f" % slope)

y = [f(v) for v in x_list]
print("x --> ", x_list)
print("y --> ", y)
derivatives = [get_derive(f, value) for value in x_list]
print("derivatives --> ", derivatives)

plt.plot(x_list,y, label='x vs y = f(x)')
plt.plot(x_list,derivatives, label='derivatives')
plt.xlabel("x")
plt.ylabel("y = f(2x)")
plt.legend()
plt.show()
```

- **Observation**
- When both x & y[f(x)] are same with multiplcation of 2 , then slope/rate of change will same always be constant. i.e., 2
